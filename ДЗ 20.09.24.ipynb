```python
# Определяем функцию для вычисления средней цены велосипеда в зависимости от количества его владельцев
```

```python
def calculate_avg_price(row):
    # Проверяем количество владельцев велосипеда и возвращаем соответствующее значение
    if row['Owner'] == '1st owner':
        return row['Ex_Showroom_Price'] * 1.1  # Увеличиваем цену на 10% для велосипедов с 1 владельцем
    elif row['Owner'] == '2nd owner':
        return row['Ex_Showroom_Price'] * 0.9   # Уменьшаем цену на 10% для велосипедов с 2 владельцами
    elif row['Owner'] == '3nd owner':
        return row['Ex_Showroom_Price'] * 0.9   # Уменьшаем цену на 20% для велосипедов с 3 владельцами
    else:
        return row['Ex_Showroom_Price']         # Если состояние не указано, возвращаем оригинальную цену
```

```python
df_Bike_Prices['discounted_price_Owner'] = df_Bike_Prices.apply(calculate_avg_price, axis=1)
```

```python
df_Bike_Prices
```

```python
# Функция для нормализации цен на велосипеды
def Normalize_price(Price):
    """Нормализует цены, деля каждую цену на максимальную цену."""
    max_Price = Price.max()
    return Price / max_Price
```

```python
df_Bike_Prices['Normalized_Price'] = df_Bike_Prices['Ex_Showroom_Price'].transform(Normalize_price)
```
