```python
def get_title_and_text(url):
  page = requests.get(url)
  if page.status_code == 200:
    soup = BeautifulSoup(page.text, "html.parser")
    title = soup.title.string
    text = ''
    for block in soup.find_all(class_ = 'detale-news-block__text'):
      if block.find('div', class_="article__article"):
        pass
      else:
        text = ''.join([text, block.get_text()])
    return title, text
  else:
    print("Ошибка при получении страницы:", page.status_code)

def clean_text(text, stopwords):
  words = [word for word in re.split(';|\.| ', text) if word not in stopwords_ru]
  text = ' '.join(words)
  return text

def evaluate_sentiment(text):
  tokenizer = RegexTokenizer()
  model = FastTextSocialNetworkModel(tokenizer=tokenizer)
  data = model.predict([text], k=5)
  return pd.DataFrame(data)

def evaluate_news(url, stopwords_ru):
  title, text = get_title_and_text(url)
  clean = clean_text(text, stopwords_ru)
  sentiment = evaluate_sentiment(clean)
  results = pd.DataFrame({'title' : title}, index = [0])
  results = pd.merge(results, sentiment, left_index = True, right_index = True)
  return results
```

```python
url = "https://newsivanovo.ru/fn_1582688.html"
stopwords_ru = stopwords.words("russian")
evaluate_news(url, stopwords_ru)
```
