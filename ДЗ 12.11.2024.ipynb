```python
# Получаем текст из указанного URL
url = "http://www.lib.ru/RUSSLIT/OSTROWSKIJ/kak_zakalyalas_stal.txt"
response = requests.get(url)
text = response.text
```

```python
# Приводим текст к нижнему регистру и разбиваем на слова
words = text.lower().split()
```

```python
# Считаем частоту слов
word_counts = Counter(words)
```

```python
# Получаем 10 самых распространенных слов
most_common_words = word_counts.most_common(10)
```

```python
# Разделяем слова и их частоты для построения гистограммы
words, counts = zip(*most_common_words)
```

```python
# Строим гистограмму
plt.bar(words, counts)
plt.title('10 самых часто встречающихся слов')
plt.xlabel('Слова')
plt.ylabel('Частота')
plt.xticks(rotation=45)
plt.show()
```
