```python
import pandas as pd
import numpy as np
```

```python
df_sales = pd.read_csv(r"https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Product%20Sales%20Data.csv")
```

```python
df_sales
```

```python
df_long = pd.melt(df_sales,
                  id_vars = 'Month', 
                  value_vars = ['Cream', 'Detergent', 'Moisturizer', 'Sanitizer', 'Shampoo', 'Soap'],
                  value_name = 'Sales',
                  ignore_index =  ['Total Units', 'Total Profit'])
```

```python
df_long
```

```python
#Посчитаем общие продажи продуктов за все месяцы, указанные в таблице. 
#По полученным данным можно увидеть спрос покупателей и выявить самые популярные и непопулярные продукты.
```

```python
pivot_table = pd.pivot_table(df_long,
                             values=['Sales'],
                             index=['variable'],
                             aggfunc={'Sales': 'sum'}) 
```

```python
pivot_table
```

```python
(
    df_long
    .groupby(['variable'])
    .agg({'Sales': [('Сумма', 'sum'), ('Cреднее', 'mean'), 'min', 'max', 'count']})
)
```
