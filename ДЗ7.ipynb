```python
mu = titanic_all['fare_filled_mean'].mean()
sigma = titanic_all['fare_filled_mean'].std()
lower_bound = mu - 3*sigma
upper_bound = mu + 3*sigma
print(f"Выбросами будут считаться все продажи, у которых цена находится за пределами интервала [{lower_bound}, {upper_bound}]")
```

```python
titanic_all[(titanic_all['fare_filled_mean']<=lower_bound)|(titanic_all['fare_filled_mean']>=upper_bound)]
```

```python
titanic_all_no_outliers = titanic_all[(titanic_all['fare_filled_mean']>=lower_bound)&(titanic_all['fare_filled_mean']<=upper_bound)]
```

```python
mu = titanic_all['age'].mean()
sigma = titanic_all['age'].std()
lower_bound = mu - 3*sigma
upper_bound = mu + 3*sigma
print(f"Выбросами будут считаться все продажи, у которых количество находится за пределами интервала [{lower_bound}, {upper_bound}]")
```

```python
titanic_all_no_outliers = titanic_all_no_outliers[(titanic_all_no_outliers['age']>=lower_bound)&(titanic_all_no_outliers['age']<=upper_bound)]
```

```python
#Нормализовать данные
```

```python
titanic_all_no_outliers['norm_price'] = (titanic_all_no_outliers['fare_filled_mean'] - titanic_all_no_outliers['fare_filled_mean'].min()) / (titanic_all_no_outliers['fare_filled_mean'].max() - titanic_all_no_outliers['fare_filled_mean'].min()) 
```
